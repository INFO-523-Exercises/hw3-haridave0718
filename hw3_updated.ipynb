```python
data['artist_gender'] = data['artist_gender'].map({'Male': 1, 'Female': 0})

# Convert the column to float
data['artist_gender'] = data['artist_gender'].astype(float)
```

```python
data['artist_race'] = data['artist_race'].apply(lambda x: "Non-White" if x != "White" else "White")
```

```python
data['artist_race'] = data['artist_race'].map({'White': 1, 'Non-White': 0})

# Convert the column to float
data['artist_race'] = data['artist_race'].astype(float)
```

```python
data['artist_ethnicity'] = data['artist_ethnicity'].map({'Hispanic or Latino origin': 1, 'Not Hispanic or Latino origin': 0})

# Convert the column to float
data['artist_ethnicity'] = data['artist_ethnicity'].astype(float)
```

**Decision Tree Classifier**

```python
from sklearn import tree
Y = data['book']
X = data[['artist_nationality', 'artist_gender', 'artist_race', 'artist_ethnicity', 'moma_count_to_year', 'whitney_count_to_year']]

# Create a decision tree classifier object with entropy as the impurity measure and a maximum depth of 3
clf = tree.DecisionTreeClassifier(criterion='entropy', max_depth=3)

# Train the classifier on the labeled data
clf = clf.fit(X, Y)
```

```python
import pydotplus
from IPython.display import Image

dot_data = tree.export_graphviz(clf, feature_names=X.columns, class_names=['Janson', 'Gardner'], filled=True, out_file=None)

# Create the decision tree graph
graph = pydotplus.graph_from_dot_data(dot_data)

# Display the decision tree graph
Image(graph.create_png())
```

```python
testData = [['Cristiano Ronaldo',0,0,0,0,1,1,'Gardner'],
           ['Lionel Messi',1,0,0,0,1,1,'Janson'],
           ['Neymar Jr',1,0,0,1,1,0,'Gardner'],
           ['Kylian Mbappe',1,1,1,0,0,0,'Janson']]
testData = pd.DataFrame(testData, columns=data.columns)
testData
```

```python
testY = testData['book']
testX = testData.drop(['artist_name', 'book'], axis=1)

# Make predictions using the decision tree classifier
predY = clf.predict(testX)

# Create a DataFrame with artist names and predicted classes
predictions = pd.DataFrame({'artist_name': testData['artist_name'], 'Predicted_book': predY})

# Display the predictions
predictions
```

```python
from sklearn.metrics import accuracy_score

print('Accuracy on test data is %.2f' % (accuracy_score(testY, predY)))
```

Model Overfitting

```python
#########################################
# Training and Test set creation
#########################################

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.8, random_state=1)
import numpy as np
from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

#########################################
# Model fitting and evaluation
#########################################

maxdepths = [2,3,4,5,6,7,8,9,10,15,20,25,30,35,40,45,50]

trainAcc = np.zeros(len(maxdepths))
testAcc = np.zeros(len(maxdepths))

index = 0
for depth in maxdepths:
    clf = tree.DecisionTreeClassifier(max_depth=depth)
    clf = clf.fit(X_train, Y_train)
    Y_predTrain = clf.predict(X_train)
    Y_predTest = clf.predict(X_test)
    trainAcc[index] = accuracy_score(Y_train, Y_predTrain)
    testAcc[index] = accuracy_score(Y_test, Y_predTest)
    index += 1

#########################################
# Plot of training and test accuracies
#########################################

plt.plot(maxdepths,trainAcc,'ro-',maxdepths,testAcc,'bv--')
plt.legend(['Training Accuracy','Test Accuracy'])
plt.xlabel('Max depth')
plt.ylabel('Accuracy')
```

*   The dataset is split into training and test sets (80% training, 20% test) using 'train_test_split.'
*   A decision tree classifier is trained with various maximum depths.
The code computes and plots training and test accuracies against different maximum depths.
*   The plot helps visualize how the decision tree's performance changes with depth, showing that increasing depth can improve training accuracy but may lead to overfitting, reducing test accuracy. The goal is to find the depth that maximizes test accuracy while avoiding overfitting.





K-nearest neighbor classifier

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

numNeighbors = [1, 5, 10, 15, 20, 25, 30]
trainAcc = []
testAcc = []

for k in numNeighbors:
    clf = KNeighborsClassifier(n_neighbors=k, metric='minkowski', p=2)
    clf.fit(X_train, Y_train)
    Y_predTrain = clf.predict(X_train)
    Y_predTest = clf.predict(X_test)
    trainAcc.append(accuracy_score(Y_train, Y_predTrain))
    testAcc.append(accuracy_score(Y_test, Y_predTest))
```

```python
plt.plot(numNeighbors, trainAcc, 'ro-', numNeighbors, testAcc,'bv--')
plt.legend(['Training Accuracy','Test Accuracy'])
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
```

The plot helps in understanding how the choice of the number of neighbors affects the KNN model's performance.

Generally, as the number of neighbors increases, the model may become more robust but less sensitive to local variations in the data.

The goal is to find a balance where the model generalizes well to new data (test accuracy is high) without overfitting to the training data (training accuracy is not excessively high).

The plot allows for the identification of an optimal number of neighbors that maximizes test accuracy.

**Linear Classifier**

```python
from sklearn import linear_model
from sklearn.svm import SVC

C = [0.01, 0.1, 0.2, 0.5, 0.8, 1, 5, 10, 20, 50]
LRtrainAcc = []
LRtestAcc = []
SVMtrainAcc = []
SVMtestAcc = []

for param in C:
    clf = linear_model.LogisticRegression(C=param)
    clf.fit(X_train, Y_train)
    Y_predTrain = clf.predict(X_train)
    Y_predTest = clf.predict(X_test)
    LRtrainAcc.append(accuracy_score(Y_train, Y_predTrain))
    LRtestAcc.append(accuracy_score(Y_test, Y_predTest))

    clf = SVC(C=param,kernel='linear')
    clf.fit(X_train, Y_train)
    Y_predTrain = clf.predict(X_train)
    Y_predTest = clf.predict(X_test)
    SVMtrainAcc.append(accuracy_score(Y_train, Y_predTrain))
    SVMtestAcc.append(accuracy_score(Y_test, Y_predTest))
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))
ax1.plot(C, LRtrainAcc, 'ro-', C, LRtestAcc,'bv--')
ax1.legend(['Training Accuracy','Test Accuracy'])
ax1.set_xlabel('C')
ax1.set_xscale('log')
ax1.set_ylabel('Accuracy')

ax2.plot(C, SVMtrainAcc, 'ro-', C, SVMtestAcc,'bv--')
ax2.legend(['Training Accuracy','Test Accuracy'])
ax2.set_xlabel('C')
ax2.set_xscale('log')
ax2.set_ylabel('Accuracy')
```

Logistic Regression (LR) and SVM models are trained and tested for various 'C' values.

Training and test accuracies for both LR and SVM are recorded.

Plots display how accuracy changes with 'C' values on a logarithmic scale.

The goal is to identify the 'C' value that optimizes test accuracy while avoiding overfitting on the training set.

Ensemble Methods

```python
from sklearn import ensemble
from sklearn.tree import DecisionTreeClassifier

numBaseClassifiers = 500
maxdepth = 10
trainAcc = []
testAcc = []

clf = ensemble.RandomForestClassifier(n_estimators=numBaseClassifiers)
clf.fit(X_train, Y_train)
```

```python
Y_predTrain = clf.predict(X_train)
Y_predTest = clf.predict(X_test)
trainAcc.append(accuracy_score(Y_train, Y_predTrain))
testAcc.append(accuracy_score(Y_test, Y_predTest))

clf = ensemble.BaggingClassifier(DecisionTreeClassifier(max_depth=maxdepth),n_estimators=numBaseClassifiers)
clf.fit(X_train, Y_train)
```

```python
Y_predTrain = clf.predict(X_train)
Y_predTest = clf.predict(X_test)
trainAcc.append(accuracy_score(Y_train, Y_predTrain))
testAcc.append(accuracy_score(Y_test, Y_predTest))

clf = ensemble.AdaBoostClassifier(DecisionTreeClassifier(max_depth=maxdepth),n_estimators=numBaseClassifiers)
clf.fit(X_train, Y_train)
```

```python
Y_predTrain = clf.predict(X_train)
Y_predTest = clf.predict(X_test)
trainAcc.append(accuracy_score(Y_train, Y_predTrain))
testAcc.append(accuracy_score(Y_test, Y_predTest))

methods = ['Random Forest', 'Bagging', 'AdaBoost']
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))
ax1.bar([1.5,2.5,3.5], trainAcc)
ax1.set_xticks([1.5,2.5,3.5])
ax1.set_xticklabels(methods)
ax2.bar([1.5,2.5,3.5], testAcc)
ax2.set_xticks([1.5,2.5,3.5])
ax2.set_xticklabels(methods)
```

Compare the performance of Random Forest, Bagging, and AdaBoost ensemble methods, each with 500 Decision Tree base classifiers of maximum depth 10.

The models are trained and evaluated on training and test sets.

Bar plots illustrate the accuracy of each ensemble method.

The goal is to identify the method with the highest test accuracy, offering insights into their effectiveness.
